```python
#결측값을 확인 했으므로 4분위를 통해서 이상값 제거
q1 = train_set["Fare"].quantile(q=0.25)
q3 = train_set["Fare"].quantile(q=0.75)
IQR = q3-q1
IQR

# 1.5범위를 지정하여 결측값을 제거한 데이터를 data_IQR에 따로 저장 
train_set = train_set[(train_set["Fare"] < q3 + IQR * 1.5) & (train_set["Fare"] > q1 - IQR * 1.5)]
train_set

#결측값을 확인 했으므로 4분위를 통해서 이상값 제거
q1 = test_set["Fare"].quantile(q=0.25)
q3 = test_set["Fare"].quantile(q=0.75)
IQR = q3-q1
IQR

# 1.5범위를 지정하여 결측값을 제거한 데이터를 data_IQR에 따로 저장 
test_set = test_set[(train_set["Fare"] < q3 + IQR * 1.5) & (test_set["Fare"] > q1 - IQR * 1.5)]
test_set
```
